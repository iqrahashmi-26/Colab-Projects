```python
airline_1= pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
airline_1
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose  # tsa = Time Series ANalysis
```

```python
result = seasonal_decompose(airline_1['Thousands of Passengers'], model='multiplicative')
result.plot();
```

```python
result_1= seasonal_decompose(airline_1['Thousands of Passengers'], model='add')
result_1.plot();
```

```python
airline_1['6-months-SMA']= airline_1['Thousands of Passengers'].rolling(window=6).mean()
airline_1['12-months-SMA']= airline_1['Thousands of Passengers'].rolling(window=12).mean()
airline_1['120-months-SMA']= airline_1['Thousands of Passengers'].rolling(window=120).mean()
airline_1['240-months-SMA']= airline_1['Thousands of Passengers'].rolling(window=240).mean()
```

```python
airline_1.plot()
```

```python
airline_1['EWMA12']= airline_1['Thousands of Passengers'].ewm(span=12,adjust=False).mean()
airline_1['EWMA120']= airline_1['Thousands of Passengers'].ewm(span=120,adjust=False).mean()
airline_1['EWMA1200']= airline_1['Thousands of Passengers'].ewm(span=1200,adjust=False).mean()
airline_1['EWMA2400']= airline_1['Thousands of Passengers'].ewm(span=2400,adjust=False).mean()
```

```python
airline_1.plot()
```
